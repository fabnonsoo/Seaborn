## Regression Plots - lmplot()
Seaborn has many built-in capabilities for regression plots, but we will only cover the lmplot() function for now.

lmplot allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Ex 1i:- Here, we have a scatter plot with a linear fit on top of it...

sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# Ex 2:- Using "hue" to show 2 scatter and linear fits for different categories by colors....

sns.lmplot(x='total_bill',y='tip',data=tips, hue="sex")
```

```python
# Using "markers" to show 2 scatter and linear fits for different categories....


# Ex 3:-

sns.lmplot(x='total_bill',y='tip',data=tips, hue="sex", markers=["x", "o"])
```

```python
# For instance, if you cannot see your "markers" & want to make them bigger, we specify the scatter plot args: "scatter_kws"
# "s" in "scatter_kws" means -> size of scatter plot

# Ex 4:-

sns.lmplot(x='total_bill',y='tip',data=tips, hue="sex", markers=["x", "o"], scatter_kws={"s":75})
```

### Using Grid (columns) instead of seperating by "hue"

```python
# Ex 5:-

sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# Using grid (col) instead of hue to seperate categories into columns....

sns.lmplot(x='total_bill',y='tip',data=tips, col="sex")
```

```python
# Ex 6:-

# Using grid (col & row) instead of hue to seperate categories into columns....

sns.lmplot(x='total_bill',y='tip',data=tips, col="sex", row="time")
```

```python
# Ex 7:- Using hue and column grid....

sns.lmplot(x='total_bill',y='tip',data=tips, col="day", hue="sex")
```

### Aspect and Size
Continuation from Example 7: Seaborn figures can have their size and aspect ratio adjusted with the size and aspect parameters.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', aspect=0.6, size=8)
```
