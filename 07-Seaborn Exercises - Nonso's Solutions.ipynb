### Q1

```python
# x and y --> are column names from the seaborn DataFrame, WHILE data --> is the name of the DataFrame...
# To include outlines in the distplot() -> we use the marginal_kws, kde_kws and hist_kws..


sns.jointplot(x="fare",y="age",data=titanic, marginal_kws={"color":"purple", "kde_kws":{"color":"crimson","lw":2}, "hist_kws": {"ec":"black","lw":2}})
```

### Q2?

```python
# Adding --> ".annotate(stats.pearsonr)" to show the pearson coefficient and p-value on the chart...
#  Also --> import scipy.stats as stats

import scipy.stats as stats
sns.jointplot(x="fare", y="age",data=titanic, marginal_kws={"color":"purple", "kde_kws":{"color":"crimson","lw":2}, "hist_kws": {"ec":"black","lw":2}}).annotate(stats.pearsonr)
```

### Q3?

```python
sns.distplot(titanic["fare"], kde=False, bins=30, hist_kws=dict(ec="black", linewidth=2))
```

### Q4?

```python
sns.boxplot(x="class", y="age", data=titanic)
```

### Q5?

```python
# Combining the stripplot versus the violinplot creates a swarmplot...
# "swarmplot" is very similar to a "stripplot", except the points are adjusted so that they don't overlap...
# The drawback of a swarmplot is that it doesn't scale that well to very large numbers - It's advisable not to use swarmplot for very large datasets.. 

sns.swarmplot(x="class", y="age", data=titanic)
```

### Q6?

```python
sns.countplot(data=titanic,x="sex")
```

### Q7?

```python
sns.heatmap(titanic.corr(), cmap="coolwarm")
```

### Q8?

```python
fg = sns.FacetGrid(data=titanic, col="sex")
fg.map(plt.hist,"age")
```
