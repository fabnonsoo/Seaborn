## Style and Color
Lets show how to control figure aesthetics in seaborn formally.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Styles
Setting particular styles.

```python
# Ex 1:-

sns.countplot(data=tips,x="sex")
```

```python
# Ex 2:- 

sns.set_style('darkgrid')
sns.countplot(data=tips,x="sex")
```

```python
sns.set_style('darkgrid')
sns.countplot(data=tips,x="sex", hue="sex")
```

### Spine Removal

```python
sns.set_style('ticks')
sns.countplot(data=tips,x="sex")
```

```python
# Ex 3i:- Removing the top and right spines

sns.set_style('ticks')
sns.countplot(data=tips,x="sex")
sns.despine()
```

```python
# Ex 3ii:- Removing the remaining spines -left and bottom...

sns.set_style('ticks')
sns.countplot(data=tips,x="sex")
sns.despine(left=True, bottom=True)
```

### Aspect and Size
You can use matplotlib's *plt.figure(figsize=(width,height) * to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect.

```python
# Ex 4i:- Non Grid Plot

plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Ex 4ii:- Grid Type Plot

sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

### Scale and Conntext
The set_context() allows you to override default parameters.

```python
# Ex 5i:- 

sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
# Ex 5ii:- Adding the font_scale = 2, makes the figure 2 times its default size...

sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

```python
# Ex 5iii:- changing the context to "notebook"..

sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```

```python
# Ex 6:- Adding matplotlib color map...

sns.set_context('notebook')
sns.lmplot(data=tips, x='total_bill', y='tip', hue="sex",palette="seismic")
```
