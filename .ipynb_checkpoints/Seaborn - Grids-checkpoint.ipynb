## Grids
Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris["species"].unique()
```

### pairplot
pairplot is a simpler version of PairGrid.

```python
# Ex 1i:- 

sns.pairplot(iris)
```

```python
sns.pairplot(iris, diag_kws={"edgecolor":"black"})
```

### PairGrid
Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# You can see that from the chart, a PairGrid just gives an empty chart without data points...
# Which makes us have more control as to the type of chart we want in our grid.

# Ex 2i:-

sns.PairGrid(iris)
```

```python
# Now lets map a scatter plot to the grid 

# Ex 2ii:-

scg = sns.PairGrid(iris)
scg.map(plt.scatter)
```

```python
# mapping scatter, distplot & kdeplot respectively to the diagonals (upper diagonal, diagonal & lower diag)
# From the charts, you can see that we use the PairGrid to get whatever kinda plot we want in our grids by mapping....

# Ex 2iii:-

scg = sns.PairGrid(iris)
scg.map_diag(sns.distplot)
scg.map_upper(plt.scatter)
scg.map_lower(sns.kdeplot)
```

```python
# including edegecolor for Ex 2iii...

scg = sns.PairGrid(iris)
scg.map_diag(sns.distplot, hist_kws=dict(ec="black", linewidth=2))
scg.map_upper(plt.scatter)
scg.map_lower(sns.kdeplot)
```

### FacetGrid
FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Ex 3i:-

fg = sns.FacetGrid(data=tips, col="time", row="smoker")
```

```python
# Ex 3ii:- Mapping 1 argument to plot type "distplot" for the FacetGrid. With the x-axis representing the "total_bill"..

fg = sns.FacetGrid(data=tips, col="time", row="smoker")
fg.map(sns.distplot,"total_bill")
```

```python
# Including edgecolor for Ex. 3ii

fg = sns.FacetGrid(data=tips, col="time", row="smoker")
fg.map(sns.distplot,"total_bill", hist_kws=dict(ec="black", linewidth=2))
```

```python
# Ex 3ii:- Mapping 2 arguments to the plot type "scatter". With the x-axis representing the "total_bill" & y-axis: tip..

fg = sns.FacetGrid(data=tips, col="time", row="smoker")
fg.map(plt.scatter,"total_bill", "tip")
```
