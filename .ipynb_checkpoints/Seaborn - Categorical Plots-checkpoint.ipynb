## Categorical Plots

With categorical plots we're going to be concerned about seeing the distributions of a categorical column such as "sex", "smoker", "day" and "time" as seen in our dataset.

There are a few main plot types for using seaborn to plot categorical data! 

    * factorplot
    * boxplot
    * violinplot
    * stripplot
    * swarmplot
    * barplot
    * countplot

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### barplot

The barplot is a general plot that allows us to aggregrate categorical data based off some functions such as the mean.

```python
# Here, we use the x-axis as non-numerical/categorical & y-axis as numerical column
# The results of the plot shows the mean of total_bill per categorical column value. 
# So, the average total_bill for Male is going to be slightly higher than that of the Female...

# Ex 1:- 

sns.barplot(x="sex", y="total_bill", data=tips)
```

Lets change the estimator object to own function, that converts a vector to a scalar:

```python
# Ex 2:- Figuring out the standard deviation per gender. i.e Do Female's have more variant in total bills than Male's

import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot
This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value.

```python
# The result of this chart shows that they're more Males in the dataset than Females...

# Ex 3:- 

sns.countplot(x="sex", data=tips)
```

### boxplot

boxplots and violinplots are used to show the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
# This boxplot shows the quartiles of the dataset, while the whiskers extend to show the rest of the distribution.
# The dots shown are outliers

# Ex 4i:-

sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# The boxplot is splitted by day and smoker status - hue="smoker" 
# The 1st boxplot is for the day, WHILE, the 2nd boxplot shows the smaker column (In all of the days, nobody smokes)...
# For Fridays, people tend to have a higher total_bill if there were not a smoker versus if they're were...

# Ex 4ii:-

sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker" )
```

```python
# Ex 4iii:- Can do entire dataframe with orient='h'

sns.boxplot(data=tips,palette='rainbow',orient='h')
```

### violinplot

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
# The normal boxplot

sns.boxplot(x="day", y="total_bill", data=tips)         
```

```python
# The violinplot unlike the boxplot allows us to plot all the components that corresponds to the actual data points
# It is harder to read than a boxplot...

# Ex 5i:- The violinplot

sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# Ex 5ii:- 

sns.violinplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
# Ex 5iii:- we use split=True to split the sex column to show more information...
# The "sex" column parameter (hue="sex") must be passed
# When dealing with more experienced database professionals who are Data Scientist, you can use a violin-plot
# For CEO/Executive management you can use the boxplot.

sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split="True")
```

### stripplot

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
# Stripplot draws a scatter plot where on variable is categorical

# Ex 6i:-

sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

```python
# Ex 6ii:-

sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex")
```

```python
# The "split" parameter in stripplot has been renamed to "dodge"
# "dodge" seperates the data points into Male & Female as seen in the legend and chart...

# Ex 6iii:- 

sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex", dodge=True)
```

### swarmplot

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
# Combining the stripplot versus the violinplot creates a swarmplot...
# "swarmplot" is very similar to a "stripplot", except the points are adjusted so that they don't overlap...
# The drawback of a swarmplot is that it doesn't scale that well to very large numbers - It's advisable not to use swarmplot for very large datasets.. 

# Ex 7i:-

sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
# Ex 7ii:- Also in swarmplot, split has been renamed to dodge...

sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True)
```

### Combining Categorical Plots

```python
# This kind of plot isn't good for explaining to an executive but for exploratory data analysis...

# Ex 8:- 

sns.violinplot(x="day", y="total_bill", data=tips, palette='rainbow')
sns.swarmplot(x="day", y="total_bill", data=tips, color='black', size=3)
```

### factorplot or catplot
factorplot is the most general form of a categorical plot. It can take in a kind parameter to adjust the plot type.

```python
# Ex 9i:- Th "kind" argument is essentially what's going to define the factorplot... 

sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day',y='total_bill', data=tips, kind='bar')
```

```python
# You can use kind="violin" to call a violinplot from a factorplot/catplot
# It's just preferable to call the violinplot as -> "sns.violinplot"

# Ex 9ii:-

sns.catplot(x='day',y='total_bill', data=tips, kind='violin')
```
