### Distribution Plots
Let's see some plots that allow us to visualize the distribution of a data set. These plots are:

    * distplot
    * jointplot
    * pairplot
    * rugplot
    * kdeplot

### distplot
The distplot shows the distribution of a univariate set of observations.

```python
# The result of this code will produce a Histogram w/KDE. 
# KDE is the line that cuts across the histogram. It means a Kernel Density estimation....
# A histogram is just a distribution of where the "total_bill" lies and the bars on the x-axis are known as bins....

# Ex 2i:-

sns.distplot(tips["total_bill"])
```

```python
# Ex 2ii:- Removing the KDE from the plot...

sns.distplot(tips["total_bill"], kde=False)
```

```python
# Ex 2iii:- To view the hist more, we can increase the bin size by passsing it as an argument...

sns.distplot(tips["total_bill"], kde=False, bins=40)
```

```python
# For "seaborn" we use "hist_kws=dict" keyword argument to ensure the bins are seperated by outlines on a displot...
# For pandas we use --> df.plot(kind="hist", edgecolor="k") to seperate the bins w/edges..
# While for matplotlib plots, we can directly use the "edgecolor" or "ec" keyword argument.....

# Ex 2iv:- 

sns.distplot(tips["total_bill"], kde=False, bins=40, hist_kws=dict(ec="black", linewidth=2))
```

### jointplot
jointplot() allows you to basically match up two distplots for bivariate data (2 variables), meaning you can combine two distplots. With your choice of what kind parameter to compare with:

    * “scatter”
    * “reg”
    * “resid”
    * “kde”
    * “hex”

```python
# In this example, x and y --> are column names from the seaborn DataFrame, WHILE data --> is the name of the DataFrame...


# Ex 3i:-

sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
# Ex 3ii:- To include outlines in the distplot() -> we use the marginal_kws, kde_kws and hist_kws..

sns.jointplot(x="total_bill",y="tip",data=tips, marginal_kws={"color":"purple", "kde_kws":{"color":"crimson","lw":2}, "hist_kws": {"ec":"black","lw":2}})
```

```python
# Here, we'll be passing in the "kind" parameter that allows us to affect what goes on in the jointplot()
# In Ex 3ii, we had a scatter plot, but now we'll be passing in a hex --> Hexagon distribution representation...
# In Hexagons the points w/more data are darker (btw 10 & 20), while points w/less data is lighter... 

# Ex 3iii:-


sns.jointplot(x="total_bill",y="tip",data=tips, kind="hex", marginal_kws={"color":"purple", "kde_kws":{"color":"crimson","lw":2}, "hist_kws": {"ec":"black","lw":2}})
```

```python
# Using a reg --> regression as the plot 
# And adding --> ".annotate(stats.pearsonr)" to show the pearson coefficient and p-value on the chart...
#  Also --> import scipy.stats as stats


# Ex 3iv:- 

import scipy.stats as stats
sns.jointplot(x="total_bill",y="tip",data=tips, kind="reg", marginal_kws={"color":"purple", "kde_kws":{"color":"crimson","lw":2}, "hist_kws": {"ec":"black","lw":2}}).annotate(stats.pearsonr)
```

```python
# A kde is already included in the jointplot() on its displot() 
# To include a 2-dimensional kde, we include a kde as a parameter of "kind"
# The 2-dimensional kde shows us the points of where the data points match-up/dense the most...


# Ex 3v:- 

import scipy.stats as stats
sns.jointplot(x="total_bill",y="tip",data=tips, kind="kde").annotate(stats.pearsonr)
```

### pairplot
pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

```python
# Ex 4i:- You could see that the pairplot() plots all the data for each column..

sns.pairplot(tips)
```

```python
# To include outlines on the histogram edges in the pairplot(), we use --> "diag_kws"

# Ex 4ii:-

sns.pairplot(tips, diag_kws={"edgecolor":"black"})
```

```python
# We use the "hue" argument to pass in categorical data from our DataFrame (i.e data like "sex" column - w/two categories --> Male & Female)
# The 2 colors signifies both Male and Female data points....

# Ex 4iii:-

sns.pairplot(tips, hue="sex")
```

```python
# To change the pairplot in Ex 4iii to a Kind of plot for the diagonal subplots (histogram, etc..). The default depends on whether "hue" is used or not.


# Ex 4iv:-

sns.pairplot(tips, hue="sex", diag_kind="hist", diag_kws={"edgecolor":"black"})
```

```python
# Using the parameter -> "palette" for the color you want....

# Ex 4v:-

sns.pairplot(tips, hue="sex", diag_kind="hist", palette="coolwarm", diag_kws={"edgecolor":"black"})
```

### rugplot
rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:


```python
# Ex 5:-

sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], hist_kws=dict(ec="black", linewidth=2))
```

### kdeplot - Comparing the rugplot and distplot together
kdeplots are Kernel Density Estimation plots. These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.

```python
# Here, the gray lines on top of the rugplot (blue dashes) is the Gaussian (Normal) distribution curve...

# Ex 6i:-

# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# Ex 6ii:- To get the kde plot we can sum these basis functions. i.e the gray lines - Normal dist curve.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```
